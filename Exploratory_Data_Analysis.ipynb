# Exploratory Data Analysis

```python
data.columns
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

ax=sns.countplot(x='Gender',data=data,color='red')
plt.title("Count Plot")
plt.xlabel("Gender")
plt.ylabel("Count")
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
sales_gen = data.groupby('Gender', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

```python
ax=sns.countplot(x='Age Group',data=data,color='green',hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
#Total Amount vs Age Group

sales_age = data.groupby('Age Group', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Age Group',y='Amount',data=sales_age,color='orange')
plt.show()

```

```python
#Total Number of orders from top 10 sales
sns.set_style("whitegrid")
sns.set_palette("Set2")
sales_state = data.groupby('State', as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(5)

sns.barplot(x='State',y='Orders',data=sales_state)
plt.show()
```

```python
ax=sns.countplot(x='Shadi',data=data,hue='Gender')
plt.title("Count Plot")
sns.set(rc={'figure.figsize':(7,5)})
sns.set_style("whitegrid")            # Clean grid background
sns.set_palette("Set2")
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python


# Group data by 'Shadi' and 'Gender'
sales_gen = data.groupby(['Shadi', 'Gender'], as_index=False)['Amount'].sum()

# Plot barplot with hue
sns.barplot(x='Shadi', y='Amount', hue='Gender', data=sales_gen)
plt.title("Sales by Marital Status and Gender")
plt.show()
```
